```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = "archive"
```

```python
subfolders = [name for name in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, name))]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import random

# Snflar ve her snfta ka grnt olduunu belirleyin
class_names = []
image_paths = []

# Veri yolundaki klasrleri dolan
for class_name in os.listdir(data_path):
    class_folder = os.path.join(data_path, class_name)
    if os.path.isdir(class_folder):
        class_names.append(class_name)
        image_paths.append(os.listdir(class_folder))

# Her snftan rastgele bir grnt sein
random_images = []
for class_index, class_folder in enumerate(image_paths):
    random_image = random.choice(class_folder)
    random_image_path = os.path.join(data_path, class_names[class_index], random_image)
    random_images.append((random_image_path, class_names[class_index]))

# Grntleri alt satr ve alt stundan oluan dzenlemeyle gsterin
num_rows = 6  # Satr says
num_cols = 6  # Stun says

plt.figure(figsize=(12, 12))  # Figr boyutu
for i in range(num_rows * num_cols):
    # Tm grntleri kullanmak iin mevcut olanlardan rastgele seim yapn
    if i < len(random_images):
        image_path, class_name = random_images[i]
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatnda okur, RGB'ye evirin
        
        plt.subplot(num_rows, num_cols, i + 1)  # Grnt dzenini ayarlayn
        plt.imshow(img)  # Grnty gster
        plt.title(class_name)  # Snf adn balk olarak ekle
        plt.axis('off')  # Eksenleri gizle
    else:
        break  # Rastgele grntler biterse dur

plt.tight_layout()  # Yerleimi optimize et
plt.show()  # Grntle
```

```python

import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Klasr yolunu belirleyin
base_path = data_path

# Grnt verilerini yklemek ve yeniden boyutlandrmak iin ImageDataGenerator kullann (datagen=veri oluturucu)
datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalizasyon iin
    validation_split=0.2,  # Eitim ve dorulama verilerini ayrmak iin
    horizontal_flip=True,  # Veri artrm iin
    rotation_range=10,  # Dndrme aral
    zoom_range=0.2  # Yaknlatrma aral
)

# Eitim ve dorulama veri setlerini hazrlayn
train_generator = datagen.flow_from_directory(
    base_path,
    target_size=(128, 128),  # Grnt boyutu
    batch_size=64, # her seferinde yklenen grnt says (genellikle verisays/boyut=devir boyutu eklinde hesaplanr)
    class_mode='categorical', # snflandrma modu
    subset='training'  # Eitim verisi iin
)

validation_generator = datagen.flow_from_directory(
    base_path,
    target_size=(128, 128),
    batch_size=64,
    class_mode='categorical',
    subset='validation'  # Dorulama verisi iin
)

# Basit bir CNN modeli oluturun
model = keras.Sequential([
    layers.Conv2D(25, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.5),
    layers.Conv2D(125, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(625, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(625, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(625, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# Modeli derleyin
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Modeli eitin
epochs = 3  # Eitim epoch says
batch_size = 64  # Eitimdeki batch boyutu

model.fit(
    train_generator,  # Eitim verisi iin
    epochs=epochs,  # Epoch says
    validation_data=validation_generator,  # Dorulama verisi iin
)

# Eitim sonras dorulama veri setinde model performansn deerlendirme
validation_loss, validation_accuracy = model.evaluate(validation_generator)

print("Validation Loss:", validation_loss)
print("Validation Accuracy:", validation_accuracy)

# Confusion matrix ve classification report
y_pred = model.predict(validation_generator)
y_pred_classes = np.argmax(y_pred, axis=1)  # Snf indekslerini al
y_true = validation_generator.classes  # Gerek snflar

class_labels = list(validation_generator.class_indices.keys())  # Snf etiketleri

print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred_classes))

print("Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_labels))
```
