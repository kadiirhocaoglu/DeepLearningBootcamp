```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = "archive"
```

```python
subfolders = [name for name in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, name))]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import random

# Sınıfları ve her sınıfta kaç görüntü olduğunu belirleyin
class_names = []
image_paths = []

# Veri yolundaki klasörleri dolaşın
for class_name in os.listdir(data_path):
    class_folder = os.path.join(data_path, class_name)
    if os.path.isdir(class_folder):
        class_names.append(class_name)
        image_paths.append(os.listdir(class_folder))

# Her sınıftan rastgele bir görüntü seçin
random_images = []
for class_index, class_folder in enumerate(image_paths):
    random_image = random.choice(class_folder)
    random_image_path = os.path.join(data_path, class_names[class_index], random_image)
    random_images.append((random_image_path, class_names[class_index]))

# Görüntüleri altı satır ve altı sütundan oluşan düzenlemeyle gösterin
num_rows = 6  # Satır sayısı
num_cols = 6  # Sütun sayısı

plt.figure(figsize=(12, 12))  # Figür boyutu
for i in range(num_rows * num_cols):
    # Tüm görüntüleri kullanmak için mevcut olanlardan rastgele seçim yapın
    if i < len(random_images):
        image_path, class_name = random_images[i]
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatında okur, RGB'ye çevirin
        
        plt.subplot(num_rows, num_cols, i + 1)  # Görüntü düzenini ayarlayın
        plt.imshow(img)  # Görüntüyü göster
        plt.title(class_name)  # Sınıf adını başlık olarak ekle
        plt.axis('off')  # Eksenleri gizle
    else:
        break  # Rastgele görüntüler biterse dur

plt.tight_layout()  # Yerleşimi optimize et
plt.show()  # Görüntüle
```

```python

import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Klasör yolunu belirleyin
base_path = data_path

# Görüntü verilerini yüklemek ve yeniden boyutlandırmak için ImageDataGenerator kullanın (datagen=veri oluşturucu)
datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalizasyon için
    validation_split=0.2,  # Eğitim ve doğrulama verilerini ayırmak için
    horizontal_flip=True,  # Veri artırımı için
    rotation_range=10,  # Döndürme aralığı
    zoom_range=0.2  # Yakınlaştırma aralığı
)

# Eğitim ve doğrulama veri setlerini hazırlayın
train_generator = datagen.flow_from_directory(
    base_path,
    target_size=(128, 128),  # Görüntü boyutu
    batch_size=64, # her seferinde yüklenen görüntü sayısı (genellikle verisayısı/boyut=devir boyutu şeklinde hesaplanır)
    class_mode='categorical', # sınıflandırma modu
    subset='training'  # Eğitim verisi için
)

validation_generator = datagen.flow_from_directory(
    base_path,
    target_size=(128, 128),
    batch_size=64,
    class_mode='categorical',
    subset='validation'  # Doğrulama verisi için
)

# Basit bir CNN modeli oluşturun
model = keras.Sequential([
    layers.Conv2D(25, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.5),
    layers.Conv2D(125, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(625, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(625, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(625, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# Modeli derleyin
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Modeli eğitin
epochs = 3  # Eğitim epoch sayısı
batch_size = 64  # Eğitimdeki batch boyutu

model.fit(
    train_generator,  # Eğitim verisi için
    epochs=epochs,  # Epoch sayısı
    validation_data=validation_generator,  # Doğrulama verisi için
)

# Eğitim sonrası doğrulama veri setinde model performansını değerlendirme
validation_loss, validation_accuracy = model.evaluate(validation_generator)

print("Validation Loss:", validation_loss)
print("Validation Accuracy:", validation_accuracy)

# Confusion matrix ve classification report
y_pred = model.predict(validation_generator)
y_pred_classes = np.argmax(y_pred, axis=1)  # Sınıf indekslerini al
y_true = validation_generator.classes  # Gerçek sınıflar

class_labels = list(validation_generator.class_indices.keys())  # Sınıf etiketleri

print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred_classes))

print("Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_labels))
```
