```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('datasets/insurance.csv')
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['sex','smoker', 'region'], drop_first = True)
```

```python
y = df[['charges']]
```

```python
x = df.drop('charges', axis = 1)
```

```python
x 
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[19, 27, 0, 1, 1, 0,0,1]])
```

```python
model.predict([[19, 27.9, 0,0,1,0,0,1]])
```

```python
df_error =  pd.DataFrame()
```

```python
df_error['y'] = y
```

```python
df_error
```

```python
df_error['predict'] = model.predict(x)
```

```python
df_error
```

```python
df_error['error'] = df_error['y'] - df_error['predict']
```

```python
df_error['mse_error'] = df_error['error']**2
```

```python
df_error['mae_error'] = np.abs(df_error['error'])
```

```python
df_error
```

```python
df_error['percentage_error'] = np.abs((df_error['y'] - df_error['predict'])/df_error['y'])
```

```python
df_error
```

```python
df_mean_error = df_error.mean()
```
